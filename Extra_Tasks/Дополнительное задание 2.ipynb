```python
# ПРОДОКУМЕНТИРОВАТЬ ВСЕ ФУНКЦИИ И ПРОЦЕДУРЫ

def standartize_name(name: str) -> str:
    '''
    Стандаритизирует имя контакта
    :param name: имя контакта
    :return: стандартизированный контакт
    '''
    return name.title()


def standartize_phone(phone: str) -> str:
    '''
    Стандатизирует номер телефона
    :param phone: номер телефона
    :return: стандатизированный номер телефона
    '''
    phone_digits = ''.join(char for char in phone if char.isdigit())
    if len(phone_digits) == 10:
        return '+7' + phone_digits

    elif len(phone_digits) == 11:
        return '+7' + phone_digits[1:]

    return phone_digits


def add_contact(phone_book: dict) -> dict:
    '''
    Добавляет контакт в словарь
    :param phone_book: словарь, в который записываются имена и телефоны
    :return: копию словаря номеров и телефонов
    '''
    name = input('Ввыедите имя контакта: ')
    phone = input('Введите номер телефона: ')

    standartilezed_name = standartize_name(name)
    standartized_phone = standartize_phone(phone)

    new_phone_book = phone_book.copy()
    new_phone_book[standartilezed_name] = standartized_phone

    print('Добавление прошло успешно!')
    return new_phone_book


def delete_contact(phone_book: dict) -> dict:
    '''
    Удаляет контакт из словаря
    :param phone_book: словарь, содержащий телефоны и номера
    :return: словарь с удалённым контактом или же без изменений, если контакт не найден
    '''
    name = input('Введите имя контакта: ')
    standartized_name = standartize_name(name)

    if standartized_name in phone_book:
        new_phone_book = phone_book.copy()
        del new_phone_book[standartize_name]
        print('Контакт успешно удалён!')
        return new_phone_book
    else:
        print('Контакт не найден!')
        return phone_book


def view_contact(phone_book: dict):
    '''

    :param phone_book: словарь, содержащий телефоны и номера
    :return: None
    '''
    if not phone_book:
        print('Телефонная книга пуста.')
    else:
        for name, phone in phone_book.items():
            print(f'{name}: {phone}')


def edit_contact(phone_book: dict):
    '''
    Изменить контакт
    :param phone_book: словарь, содержащий телефоны и номера
    :return:
    '''
    name = input('Введите имя контакта: ')
    standartized_name = standartize_name(name)

    if standartized_name in phone_book:
        new_phone = input('Введите номер телефона: ')
        standartized_phone = standartize_phone(new_phone)
        new_phone_book = phone_book.copy()
        new_phone_book[standartized_name] = standartized_phone
        print('Контакт успешно изменён!')
        return new_phone_book
    else:
        print('Контакт не найден!')
        return phone_book


def menu():
    '''
    Вызывает меню
    :param: None
    :return: None
    '''
    phone_book = {}
    while True:
        print('\nВыберете функцию')
        print('1. Добавить контакт')
        print('2. Удалить контакт')
        print('3. Показать список контактов')
        print('4. Изменить номер')
        print('5. Выход')

        choice = int(input('Введите номер функции: '))

        if choice == 1:
            phone_book = add_contact(phone_book)
        elif choice == 2:
            phone_book = delete_contact(phone_book)
        elif choice == 3:
            phone_book = view_contact(phone_book)
        elif choice == 4:
            phone_book = edit_contact(phone_book)
        elif choice == 5:
            print('Выход из программы.')
            break
        else:
            print('Неверный выбор, попробуйте снова.')


menu()
```
